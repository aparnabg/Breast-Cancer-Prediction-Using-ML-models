### HISTOGRAM

```python
df=pd.read_csv("/content/data.csv")

#dividing the dataframe to 3 parts mean variables , standard error (se) variables, worst variables
df_mean = data.iloc[:, 1:11]
df_se = df.iloc[:, 12:22]
df_worst = df.iloc[:, 22:]

# Define the number of bins for histograms using Sturges' rule
bin_count = int(np.ceil(np.log2(len(df))) + 1)
print(f'Number of bins according to Sturge\'s rule: {bin_count}')

fig, axes = plt.subplots(2, 5, figsize=(16, 12), sharey=True)
fig.tight_layout(pad=3.0)
axes = axes.flatten()

for i, feature in enumerate(df_mean.columns):
    sns.histplot(ax=axes[i], data=df_mean, x=feature, bins=bin_count, kde=True, line_kws={'lw': 3})
    axes[i].set_title(f"{feature} - Mean")

plt.show()
```

```python
#histograms for each feature in the standard error dataframe
fig, axes = plt.subplots(2, 5, figsize=(16, 12), sharey=True)
fig.tight_layout(pad=3.0)
axes = axes.flatten()

for i, feature in enumerate(df_se.columns):
    sns.histplot(ax=axes[i], data=df_se, x=feature, bins=bin_count, kde=True, line_kws={'lw': 3})
    axes[i].set_title(f"{feature} - Standard Error")

plt.show()
```

```python
#histograms for each feature in the worst dataframe
fig, axes = plt.subplots(2, 5, figsize=(16, 12), sharey=True)
fig.tight_layout(pad=3.0)
axes = axes.flatten()

for i, feature in enumerate(df_worst.columns):
    sns.histplot(ax=axes[i], data=df_worst, x=feature, bins=bin_count, kde=True, line_kws={'lw': 3})
    axes[i].set_title(f"{feature} - Worst")

plt.show()
```

```python
def high_corr_pairs(corr_matrix, threshold):
    high_corr_pairs = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i, len(corr_matrix.columns)):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))
    if high_corr_pairs:
        for pair in high_corr_pairs:
            print(f"({pair[0]}, {pair[1]}): {pair[2]}")
```

In the context of categorical data, Pearson correlation coefficient aren't typically used. Because they are designed to measure linear relationships between continuous variables.

#  ANOVA


```python
X=data.drop(['target'], axis=1)
y=data['target']
from sklearn.feature_selection import SelectKBest,f_classif
selector2 = SelectKBest(f_classif, k=10)
selector2.fit(X,y)
X_new = selector2.transform(X)
```

```python
plt.figure()
bars=plt.bar(range(len(selector2.scores_)), selector2.scores_)
selected=[]
for bar in bars:
    if selector2.scores_[bars.index(bar)] > 300:
        bar.set_color('r')

plt.xticks(range(len(selector2.scores_)),X.columns, rotation='vertical')
plt.title("ANOVA F-value feature selection")
plt.show()
```

```python
X.columns
```

```python
X.columns[selector2.scores_ > 300]
```

```python

X_selected = X[X.columns[selector2.scores_ > 300]]
X_selected
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'SVM': SVC()
}

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
result={}
results = {}
for name, model in models.items():
    scores = cross_val_score(model,  X_train_scaled, y_train,cv=kfold, scoring='accuracy')
    result[name] = scores.mean()
    model.fit( X_train_scaled, y_train,)
    y_pred = model.predict(X_test_scaled)
    report = classification_report(y_test, y_pred, output_dict=True)
    results[name] = {'accuracy': scores.mean(), 'recall': report['weighted avg']['recall']}



#classification report of every model
for name, model in models.items():
    model.fit( X_train_scaled, y_train,)
    y_pred = model.predict(X_test_scaled)
    report = classification_report(y_test, y_pred)
    print(f'\nClassification Report for {name}:')
    print(report)

#finding best model
for name, score in result.items():
    print(f'{name}: {score:.4f}')
best_model = max(result, key=result.get)
print(f'\nBest Model: {best_model}')
```

```python
from sklearn.metrics import confusion_matrix

best_model_instance = models[best_model]
best_model_instance.fit(X_train_scaled, y_train)
y_pred = best_model_instance.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for Best Model')
plt.show()
```

# Data augmentation

```python
from imblearn.combine import SMOTETomek

#SMOTETomek to the training data
smote =  SMOTETomek(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_selected, y)
```

```python
target_count = y_resampled.value_counts()

pie_colors = ['#8da0cb', '#FF3333']
bar_colors = ['#8da0cb', '#FF3333']

fig, axs = plt.subplots(1, 2, figsize=(12, 6))


axs[0].pie(target_count, labels=target_count.index, autopct='%1.1f%%', startangle=90, colors=pie_colors)
axs[0].set_title('Data distribution after augmentatio (Pie Chart)')


axs[1].bar(target_count.index, target_count.values, color=bar_colors)
axs[1].set_xticks(target_count.index)
axs[1].set_xticklabels(['Benign', 'Malignant'])
axs[1].set_xlabel('Target')
axs[1].set_ylabel('Count')
axs[1].set_title('Data distribution after augmentatio (Bar Plot)')

plt.tight_layout()
plt.show()
```
